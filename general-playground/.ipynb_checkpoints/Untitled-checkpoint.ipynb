```python
#%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

import qiskit.quantum_info as qi
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere
from qiskit.visualization.utils import _bloch_multivector_data
```

```python
def plot_bloch(theta,phi):
    qc = QuantumCircuit(1,global_phase=np.pi)
    qc.ry(theta,0)
    qc.rz(-phi,0)
    sv = qi.Statevector.from_instruction(qc)
    
    width, height = plt.figaspect(1)
    figsize=(1.5*width, 1.5*height)
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(1, 1, 1, projection='3d')
    
    plot_state_qsphere(sv,figsize=figsize, ax=ax)
    
bloch_interactive = widgets.interactive(plot_bloch, 
                                        theta=widgets.FloatSlider(min=0,max=np.pi,
                                                                  step=np.pi/10,continuous_update=False), 
                                        phi=widgets.FloatSlider(min=0,max=2*np.pi,
                                                                step=np.pi/20,continuous_update=False))
bloch_interactive
```

```python
def plot_bloch(theta,phi):
    qc = QuantumCircuit(1,global_phase=np.pi)
    qc.ry(theta,0)
    qc.rz(-phi,0)
    sv = qi.Statevector.from_instruction(qc)
    
    width, height = plt.figaspect(1)
    figsize=(1.5*width, 1.5*height)
    fig = plt.figure(figsize=figsize)
    ax1 = fig.add_subplot(1, 2, 1, projection='3d')
    ax2 = fig.add_subplot(1, 2, 2, projection='3d')
    plot_state_qsphere(sv, ax=ax1)
    plot_state_qsphere(sv, ax=ax2)
    
bloch_interactive = widgets.interactive(plot_bloch, 
                                        theta=widgets.FloatSlider(min=0,max=np.pi,
                                                                  step=np.pi/10,continuous_update=False), 
                                        phi=widgets.FloatSlider(min=0,max=2*np.pi,
                                                                step=np.pi/20,continuous_update=False))
bloch_interactive
```
